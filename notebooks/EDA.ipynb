```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df["assessment_category"] = pd.cut(df["assessment_score"], bins=[0, 40, 70, 100], labels=["Low", "Medium", "High"])

fig, axes = plt.subplots(2, 2, figsize=(12, 12))

columns = ["level_of_student", "material_level", "promoted_or_skipped", "assessment_category"]
titles = ["Student Levels", "Material Levels", "Promotion Status", "Assessment Score Categories"]
colors = ["Set2", "coolwarm", "pastel", "viridis"]

for i, ax in enumerate(axes.flatten()):
    df[columns[i]].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette(colors[i]), ax=ax)
    ax.set_title(titles[i])
    ax.set_ylabel("")

plt.tight_layout()
plt.show()
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("Final Processed DataFrame:\n", df.head())
```
